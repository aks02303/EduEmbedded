```python
import os

file_names = os.listdir('data_1_folder')
mstr = []
for file_ in file_names:
    splt_lst = file_.split('_')[2:-1]
    mstr.append(splt_lst)
```

```python
mstr
```

```python
import re
n_mstr = []
for topic_ in mstr:
    tmp = []
    for word_ in topic_:
        if bool(re.search(r'\d', word_)) or word_.lower() == 'part':
            break
        elif bool(re.search(r'-', word_)):
            continue
        tmp.append(word_)
    n_mstr.append(tmp)
n_mstr
```

```python
from pprint import pprint
str_lst = [' '.join(string_) for string_ in n_mstr]
pprint(str_lst)
```

```python
len(str_lst)
```

```python
str_lst_unique = list(set(str_lst))
```

```python
from matplotlib import pyplot as plt 
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('triples.csv')
df.head(1)
```

```python
l_df = df[df['variable'] == 'l_text_topics']
con_df = df[df['variable'] == 'concept_vocab_index']
pre_df = df[df['variable'] == 'prerequisite']
lev_df = df[df['variable'] == 'level']
len(l_df.prob.values)
```

```python
l_vals = l_df.prob.values
con_vals = con_df.prob.values
pre_vals = pre_df.prob.values
lev_vals = lev_df.prob.values
plt.title('l_vals')
sns.kdeplot(l_vals, bw=0.5, )
plt.show()
plt.title('con_vals')
sns.kdeplot(con_vals, bw=0.5)
plt.show()
plt.title('pre_vals')
sns.kdeplot(pre_vals, bw=0.5)
plt.show()
plt.title('lev_vals')
sns.kdeplot(lev_vals, bw=0.5)
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler, Normalizer

l_vals_tr, con_vals_tr, pre_vals_tr, lev_vals_tr = l_vals.reshape(-1, 1), con_vals.reshape(-1, 1), pre_vals.reshape(-1, 1), lev_vals.reshape(-1, 1)

rob = RobustScaler().fit_transform(l_vals_tr)
std = StandardScaler().fit_transform(l_vals_tr)
minMax = MinMaxScaler().fit_transform(l_vals_tr)
norm = Normalizer().fit_transform(l_vals_tr)
```

```python
def show():
    l1 = [rob, std, minMax, norm]
    l2 = ['rob', 'std', 'minMax', 'norm']

    for i in range(len(l1)):
        plt.title(l2[i])
        sns.kdeplot(l1[i].reshape(-1))
        plt.show()
```

```python
show()
```

```python
rob = RobustScaler().fit_transform(con_vals_tr)
std = StandardScaler().fit_transform(con_vals_tr)
minMax = MinMaxScaler().fit_transform(con_vals_tr)
norm = Normalizer().fit_transform(con_vals_tr)

show()
```
